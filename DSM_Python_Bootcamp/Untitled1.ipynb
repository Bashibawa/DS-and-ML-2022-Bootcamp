```python
import numpy as np
import matplotlib.pyplot  as plt
#%matplotlib inline
```

$$y = 2 * x$$

```python
x = np.linspace(0,1,100)
x
```

```python
y = 2 * x
plt.plot(x, y)
plt.show()
```

```python
#to add color to the graph
plt.plot(x, y, 'r')
plt.show()
```

```python
x1 = np.linspace(-3,3,100)

y1 = np.sin(x1)
y2 = np.cos(x1)

plt.plot(x1, y1, 'r')
plt.plot(x1, y2, 'b')
plt.xlabel('x')
plt.ylabel('y')
plt.title("$y =\sin(x), y=\cos(x)$")
```

```python
#creating figures and subplots
fig =plt.figure(figsize=(12,6))
ax = fig.add_axes([0, 0, 1, 1]) 
```
